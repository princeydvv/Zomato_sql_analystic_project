## Sample SQL Preview
```sql
SELECT r.name, SUM(oi.price * oi.quantity) AS revenue
FROM restaurants r
JOIN orders o ON r.restaurant_id = o.restaurant_id
JOIN order_items oi ON o.order_id = oi.order_id
GROUP BY r.name
ORDER BY revenue DESC;
```
